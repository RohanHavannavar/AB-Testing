```python
#code for permutation resampling tests
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#pareto distribution occurs lot in nature
#first lets define sample sizes i.e n and m
n = 100
m = 100
k = 1000 #resampling value

#set random seed
np.random.seed(100)

#xc and xt
xc = np.random.pareto(a = 10,size = n)
xt = np.random.pareto(a = 8,size = m)

#observed value
m_obs = (np.percentile(xt,50) - np.percentile(xc,50))


plt.style.use("classic")
sns.distplot(xc,label = "c")
sns.distplot(xt,label = "t")
plt.grid()
plt.legend()
plt.show()

#PR test
p = np.concatenate((xc,xt))

teststats =np.zeros(k)

for i in range(k):
    np.random.shuffle(p)
    tmp_xc,tmp_xt = p[:n],p[m:]
    teststats[i] = (np.percentile(tmp_xt,50) - np.percentile(tmp_xc,50))

sns.distplot(teststats)
plt.grid()
plt.show()

num_h = np.sum(teststats > m_obs)
prob = num_h / k
print("p value is:"+str(prob))




```
